```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline

names = ["r10", "r15", "r30", "r50"]
labels = ["10%", "15%", "30%", "50%"]
source = []
for name in names:
    source.append(pd.read_csv("data/epidemic-" + name + ".csv"))
a = np.array(list(range(0,len(source[0]["infections"]))))
x = a.astype(float)

```

```python
for s, n in zip(source, labels):
    if n == "50%":
        normalization = s["infecteds"][364]

for s, n in zip(source, labels):
    plt.plot(x, s["infections"]/normalization*100, label="q="+n)

plt.title('Percentage of population infected (on that day)')     
plt.legend()
plt.show()
```

```python
for s, n in zip(source, labels):
    plt.plot(x, s["deaths"]/normalization*100, label="q="+n)

plt.title('Percentage of total population that dies')    
plt.legend()
plt.show()
```

```python
for s, n in zip(source, labels):
    plt.plot(x, s["infecteds"]/normalization*100, label="q="+n)

plt.title('Percentage of infected cells (Cumulative)')    
plt.legend()
plt.show()
```
