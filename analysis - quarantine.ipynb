```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline

names = ["q00", "q01", "q02", "q03", "q05"]
labels = ["none", "10%", "20%", "30%", "50%"]
source = []
for name in names:
    source.append(pd.read_csv("data/epidemic-" + name + ".csv"))
a = np.array(list(range(0,len(source[0]["infections"]))))
x = a.astype(float)

```

```python
for s, n in zip(source, labels):
    if n == "none":
        normalization = s["infecteds"][364]
    plt.plot(x, s["infections"]/normalization*100, label="q="+n)

plt.title('Percentage of Population Infected (on that day)')     
plt.legend()
plt.show()
```

```python
for s, n in zip(source, labels):
    plt.plot(x, s["deaths"]/normalization*100, label="q="+n)

plt.title('Percentage of Population that Dies')    
plt.legend()
plt.show()
```

```python
for s, n in zip(source, labels):
    plt.plot(x, s["infecteds"]/normalization*100, label="q="+n)

plt.title('Percentage of Population Infected (Cumulative)')    
plt.legend()
plt.show()
```
